```python
def pearson(s1, s2):
    """Take two pd.Series objects and return a pearson correlation."""
    s1_c = s1 - s1.mean()
    s2_c = s2 - s2.mean()
    return np.sum(s1_c * s2_c) / np.sqrt(np.sum(s1_c ** 2) * np.sum(s2_c ** 2))
```

```python
def get_recs(movie_name, M, num):

    import numpy as np
    reviews = []
    for title in M.columns:
        if title == movie_name:
            continue
        cor = pearson(M[movie_name], M[title])
        if np.isnan(cor):
            continue
        else:
            reviews.append((title, cor))
    
    reviews.sort(key=lambda tup: tup[1], reverse=True)
    return reviews[:num]
```

###Movie which is most Similar to 'The fault in our stars':

```python
recs = get_recs('The Fault in Our Stars', M, 10)
```

```python
#pearson correlation coefficient between (Pavithra and all other classmates)
print "Pearson Correlation coefficient of :"
a= []
for i in range(0, M.shape[0], 1): # for each row in M (range(0,23))
    cor=pearson(M.ix[6], M.ix[i])
    print "(" + M.index[6] + " , " + M.index[i] + ") :", cor
    a.append(cor)

      
```

```python
def pred_rate(x, a):
    """
    Takes two parameters 'x' and 'a'. 
    Where 'x' a list of rates given by classmates. 
    'a' is a array that has corresponding pearson correlation coefficientreturn (weights for the given rates).
    predicted rate is calculated as rate= (x[0]*a[0] + x[1]*a[1] +...+x[i]*a[i]) / (a[0]+a[1]+..+a[i]) 
    returns predicted rate 
    """
    j=0
    k=0
    for i in range(0, x.shape[0], 1): # for each row in x (range(0,23))
        if x.irow(i)>0 and a[i]>0: 
            j= j+(a[i]*x.irow(i))
            k= k+a[i]
        else:
            continue
    return j/k
```
